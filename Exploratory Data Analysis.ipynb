```python
model_data = pd.read_csv('c:/Users/ericr/Documents/Python Files/Final Project/modelDevDataset.csv')
array = model_data.values
# separate array into input and output components
X = array[:,0:13]
Y = array[:,13]

model_data.shape
```

```python
# Normalize data (length of 1)
from sklearn.preprocessing import Normalizer
from numpy import set_printoptions

scaler = Normalizer().fit(X)
normalizedX = scaler.transform(X)
# summarize transformed data
set_printoptions(precision=6)
print(normalizedX[0:5,:])
```

```python
# Feature Extraction with Univariate Statistical Tests (Chi-squared for classification)
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# feature extraction
test = SelectKBest(score_func=chi2, k=4)
fit = test.fit(X, Y)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)
# summarize selected features
print(features[0:5,:])
```

```python
# Feature Extraction with RFE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# feature extraction
model = LogisticRegression()
rfe = RFE(model, 7)
fit = rfe.fit(X, Y)
print("Num Features: {}".format(fit.n_features_)) 
print("Selected Features: {}".format(fit.support_))
print("Feature Ranking: {}".format(fit.ranking_)) 
```

```python
# Feature Importance with Extra Trees Classifier
from sklearn.ensemble import ExtraTreesClassifier

# feature extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
print(model.feature_importances_)
```

```python
model_data = pd.DataFrame(normalizedX, columns = ['KY_CD', 'PD_CD', 'CRM_ATPT_CPTD_CD', 'LAW_CAT_CD', 'LOC_OF_OCCUR_DESC', 
               'PREM_TYP_DESC', 'JURISDICTION_CODE', 'SUSP_AGE_GROUP', 'SUSP_RACE', 'SUSP_SEX', 
               'VIC_AGE_GROUP', 'VIC_RACE', 'VIC_SEX'])
model_data = model_data[['PREM_TYP_DESC', 'PD_CD', 'VIC_AGE_GROUP', 'SUSP_AGE_GROUP', 
                         'VIC_RACE', 'SUSP_RACE', 'KY_CD', 'SUSP_SEX', 'VIC_SEX', 
                         'LOC_OF_OCCUR_DESC', 'JURISDICTION_CODE', 'LAW_CAT_CD', 'CRM_ATPT_CPTD_CD']]
print(model_data.shape)
```

```python
# Cross Validation Classification Accuracy
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

array2 = model_data.values
X2 = array2[:,0:13]
kfold = KFold(n_splits=10, random_state=7)
model = LogisticRegression()
scoring = 'accuracy'
results = cross_val_score(model, X2, Y, cv=kfold, scoring=scoring)
print("Accuracy: {} ({})".format(results.mean(), results.std())) 
```

```python
# Compare Algorithms
from matplotlib import pyplot
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# prepare models
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))

# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = KFold(n_splits=10, random_state=7)
    cv_results = cross_val_score(model, X2, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
   
# boxplot algorithm comparison
fig = pyplot.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
pyplot.boxplot(results)
ax.set_xticklabels(names)
pyplot.show()
```
